### Python
#### 변수와 메모리
- 변수 : 특정 값을 저장하는 메모리 공간의 이름

### Lab 1 :  자료형과 기본 연산 
#### 동적 타이핑(Dynamic Typing) 
- 변수의 메모리 공간 확보 및 자료형(정수(integer), 실수(float))을 프로그램 실행 시점에 파이썬 인터프리터가 결정
- 변수에 할당하는 메모리 공간도 값의 크기(자료 유형)에 따라 동적으로 할당받음


`
a = 10
b = 10.5
c = a + b
print(c, type(a), type(b), type(c))
`

```python
# 변수의 자료형에 따라 메모리 공간을 동적으로 할당 받음 
import sys
a = 10
b = 10.5
c = a+b
print(a, type(a), type(b), type(c), sys.getsizeof(a), sys.getsizeof(b))
```

### Lab 6 : 리스트 자료형의 메모리 구조 이해
- 관련된 나열 항목을 관리
- 리스트의 원소(콤마로 구분된 항목)는 같은 자료형(정수, 실수, 문자열, 리스트)일 필요가 없다.
- 항목의 순서는 의미가 있고, 항목 자료 값은 중복되도 산관 없다.
- 항목을 참조하기 위한 색인(index)은 0부터 시작하고, 1씩 증가
- 마지막 색인은 역순으로 -1부터 시작하여, 반대로 1씩 감소
- 첨자의 범위는 0 ~ [len(리스트 변수)-1], 또는 -1 ~ [-len(리스트 변수] 가능

```python
# 리스트 변수 a와 b를 선언하고, 리스트 변수 a와 b의 주소 값을 출력
a = [5, 4, 3, 2, 1]
b = [1, 2, 3, 4, 5]
print(id(a), id(b), a, b)
```

```python
# 리스트 변수 b에 리스트 변수 a의 주소 값을 할당
# 리스트 변수 a와 b의 주소 값은 동일한 것을 확인
a = [5, 4, 3, 2, 1]
b = [1, 2, 3, 4, 5]
b = a
print(id(a), id(b), a, b)
```

```python
# 리스트 변수 a와 b를 선언하고, 리스트 변수 a와 b의 주소 값을 출력
a = [5, 4, 3, 2, 1]
b = [1, 2, 3, 4, 5]
print(id(a), id(b), a, b)

# 리스트 변수 b에 리스트 변수 a의 주소 값을 할당
b = a
print(id(a), id(b), a, b)

# 리스트 b의 첫 번째 원소 값을 수정하면, 리스트 a의 값도 변경된다.
b[0] = 10
print(id(a), id(b), a, b)
```

```python
# 리스트 변수 a와 b를 선언하고, 리스트 변수 a와 b의 주소 값을 출력
a = [5, 4, 3, 2, 1]
b = [1, 2, 3, 4, 5]
print(id(a), id(b), a, b)

# 리스트 변수 b에 리스트 변수 a의 주소 값을 할당
b = a
print(id(a), id(b), a, b)

# 새로운 리스트 메모리를 할당하고, 그 메모리 주소를 변수 b에 할당
b = [6, 7, 8, 9, 10]
print(id(a), id(b), a, b)
```

### Lab 7 : 리스트 자료형의 메모리 구조 이해

```python
# 리스트를 메모리에 할당하고, 그 메모리 주소 값을 변수 x에 할당
# 리스트 변수 x의 주소 값을 y의 주소 값으로 할당
x = [5, 4, 3, 2, 1]
y = x
print(id(x), id(y), x, y)

# 리스트 변수 y[1] 값을 99로 변경 후, 리스트 x, y 값 출력
y[1] = 99
print(x, y)
```

```python
a = [6, 7, ['x','y'], 9, 10]

# 리스트를 다른 리스트 메모리로 복사하는 copy() 메서드를 활용하여 복사
# 리스트 변수 a, b의 메모리 주소 값이 다른 것을 확인
b = a.copy()
print(id(a), id(b), a, b)

# 복사한 리스트 b[1] 값을 99로 수정한 후, 리스트 a와 b의 값 변화를 관찰
b[1] = 99
print(a, b)
```

### Lab 8 : 리스트 자료형의 메모리 구조 이해

```python
v1 = ['a', 'b', 'c']
v2 = ['x', 'y', 'z']
print(id(v1), id(v2))

# v1의 주소 값이 변경되는 것을 관찰하고, 과정을 설명해 보자
v1 = v1 + v2
print(id(v1), id(v2), v1)
```

```python
v1 = ['a', 'b', 'c']
print(id(v1), v1)

# append() 메서드를 활용하여 요소 값을 추가한 후, 리스트 변수 v1의 주소 값 변화를 관찰
v1.append('x')
print(id(v1), v1)
```
